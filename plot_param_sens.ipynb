```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_param_sens(acc:list[float], label:list, color):
    assert len(acc) == len(label)
    label = [str(l) for l in label]
    sns.lineplot(x=label, y=acc, color=color, linestyle='--', marker='o')

plt.figure(figsize=(14,5))
plt.subplot(121)
plt.ylim(55,105)
plot_param_sens([75.35,80.56,86.81,85.07],[64,128,256,512],'r') # arousal
plot_param_sens([77.08,78.82,84.72,80.56],[64,128,256,512],'b') # valence
plot_param_sens([80.00,84.29,87.14,81.43],[64,128,256,512],'g') # stress
plot_param_sens([78.33,81.28,81.86,80.26],[64,128,256,512],'orange') # wg
plot_param_sens([90.49,93.06,94.66,92.86],[64,128,256,512],'cyan') # dsr
plot_param_sens([61.40,62.68,61.25,61.97],[64,128,256,512],'purple') # nback
# plt.legend(['arousal','valence','stress','WG','DSR','N-back'])

# plt.figure(figsize=(6.5,5))
plt.subplot(122)
plt.ylim(55,105)
plot_param_sens([81.94,82.29,86.81,80.56,83.33],[2,2.5,3,5,7.5],'r') # arousal
plot_param_sens([83.68,81.94,84.72,76.74,77.43],[2,2.5,3,5,7.5],'b') # valence
plot_param_sens([78.57,86.57,87.14,80.00,77.14],[2,2.5,3,5,7.5],'g') # stress
plot_param_sens([81.03,81.86,78.01],[2,2.5,5],'orange') # wg
plot_param_sens([93.06,94.66,92.41],[2,2.5,5],'cyan') # dsr
plot_param_sens([62.39,62.68,57.98],[2,2.5,5],'purple') # dsr
plt.legend(['arousal','valence','stress','WG','DSR','N-back'])
```
