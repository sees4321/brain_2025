```python
from utils import *
import torch

total = 0
# model = SyncNet2([7,7680],[26,371],use_lstm=True).to(DEVICE)
model = SyncNet_ablation([7,7680],[26,371]).to(DEVICE)
# model = EF_net(1).to(DEVICE)
# model = STANet().to(DEVICE)
for i in range(100):
    x = torch.randn((16,7,7680)).to(DEVICE)
    y = torch.randn((16,26,371)).to(DEVICE)
    # x = torch.randn((16,1,16,16,7680)).to(DEVICE)
    # y = torch.randn((16,2,16,16,371)).to(DEVICE)
    start = torch.cuda.Event(enable_timing=True)
    end = torch.cuda.Event(enable_timing=True)
    start.record()
    model(x,y)
    end.record()
    torch.cuda.synchronize()
    total += start.elapsed_time(end)
    print(start.elapsed_time(end))
print(f'inference time = {total/100:.2f}')
```
