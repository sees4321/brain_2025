```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_param_sens(acc:list[float], label:list):
    assert len(acc) == len(label)
    label = [str(l) for l in label]
    plt.figure(figsize=(6.5,5))
    sns.lineplot(x=label, y=acc, color='r', linestyle='--', marker='o')
    plt.ylim(55,105)

plot_param_sens([75.35,80.56,86.81,85.07],[64,128,256,512])
```
